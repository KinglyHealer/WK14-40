```python
# %% [markdown]
# # 学生成绩数据分析
# 读取C程序输出日志，数据解析、统计指标、可视化绘图
# %%
import pandas as pd
import matplotlib.pyplot as plt
import re
import os

# 设置图片中文不乱码
plt.rcParams["font.family"] = ["SimHei", "WenQuanYi Micro Hei"]
plt.rcParams["axes.unicode_minus"] = False

# %%
# 1.读取C程序输出文件
file_path = "../output/run_result.txt"
with open(file_path, "r", encoding="utf-8") as f:
    content = f.read()

# %%
# 2.正则提取姓名、分数数据
pattern = r"(\d+\.\s+)(.+?):\s*(\d+)分"
res = re.findall(pattern, content)
name_list = [i[1] for i in res]
score_list = [int(i[2]) for i in res]

# 构造DataFrame
df = pd.DataFrame({
    "姓名": name_list,
    "成绩": score_list
})
print("原始成绩数据表：")
print(df)

# %%
# 3.基础统计
score_arr = df["成绩"]
stats_dict = {
    "平均分": round(score_arr.mean(), 2),
    "最高分": score_arr.max(),
    "最低分": score_arr.min(),
    "总人数": len(df),
    "总分": score_arr.sum()
}
print("\n====成绩统计汇总====")
for k, v in stats_dict.items():
    print(f"{k}:{v}")

# %%
# 4.绘制柱状图：学生成绩分布
plt.figure(figsize=(10, 5))
plt.bar(df["姓名"], df["成绩"], color="#4285F4")
plt.title("学生成绩柱状分布图")
plt.ylabel("分数")
plt.xlabel("学生姓名")
plt.grid(axis="y", alpha=0.3)
plt.savefig("score_bar.png", dpi=300, bbox_inches="tight")
plt.show()

# %% [markdown]
# # 数据分析结论
# 1.通过图表可以直观看到每位学生分数高低，排序和C语言程序输出排名一致；
# 2.统计数据和C程序内部计算的平均分、极值做交叉校验，验证C代码计算无计算错误；
# 3.可快速筛选不及格(<60)、优秀(>=90)分段学生，便于成绩分级。
```
